```python
# ==========================================
# STEP 2: Load Dataset
# ==========================================

# Load CSV file
df = pd.read_csv("Mall_Customers.csv")

# Display first 5 rows
print("First Five Rows")
print(df.head())

# Dataset Information
print("\nDataset Information")
print(df.info())

# Dataset Shape
print("\nDataset Shape")
print(df.shape)
```

```python
print("Missing Values in Each Column")
print(df.isnull().sum())
```

```python
numerical_columns = [
    "Age",
    "Annual Income (k$)",
    "Spending Score (1-100)"
]

for column in numerical_columns:
    df[column].fillna(df[column].mean(), inplace=True)

print("Numerical Missing Values Filled")
```

```python
df["Gender"].fillna(df["Gender"].mode()[0], inplace=True)

print("Categorical Missing Values Filled")
```

```python
print("Missing Values After Filling")
print(df.isnull().sum())
```

```python
encoder = LabelEncoder()

df["Gender"] = encoder.fit_transform(df["Gender"])

print(df.head())

print("\nEncoded Values")
print(df["Gender"].unique())
```

```python
encoder = LabelEncoder()

df["Gender"] = encoder.fit_transform(df["Gender"])

print(df.head())

print("\nEncoded Values")
print(df["Gender"].unique())
```

```python
# Annual Income

Q1 = df["Annual Income (k$)"].quantile(0.25)
Q3 = df["Annual Income (k$)"].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

print("Lower Limit:", lower)
print("Upper Limit:", upper)

# Spending Score

Q1_score = df["Spending Score (1-100)"].quantile(0.25)
Q3_score = df["Spending Score (1-100)"].quantile(0.75)

IQR_score = Q3_score - Q1_score

lower_score = Q1_score - 1.5 * IQR_score
upper_score = Q3_score + 1.5 * IQR_score

print("Lower Limit:", lower_score)
print("Upper Limit:", upper_score)

# Remove Outliers

df = df[
    (df["Annual Income (k$)"] >= lower) &
    (df["Annual Income (k$)"] <= upper) &
    (df["Spending Score (1-100)"] >= lower_score) &
    (df["Spending Score (1-100)"] <= upper_score)
]

print("Dataset Shape After Removing Outliers")
print(df.shape)
```

```python
scaler = StandardScaler()

columns_to_scale = [
    "Age",
    "Annual Income (k$)",
    "Spending Score (1-100)"
]

df[columns_to_scale] = scaler.fit_transform(df[columns_to_scale])

print(df.head())
```

```python
print("Mean")
print(df[columns_to_scale].mean())

print("\nStandard Deviation")
print(df[columns_to_scale].std())
```

```python
df.to_csv("Mall_Customers_Processed.csv", index=False)

print("Processed Dataset Saved Successfully")
```
