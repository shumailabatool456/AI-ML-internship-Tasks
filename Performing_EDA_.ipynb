```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(42)
```

```python
employees = ["A", "B", "C", "D", "E", "F"]

data = {
    "Employee": employees,
    "Age": np.random.randint(22, 51, size=6),
    "Salary": np.random.randint(30000, 100001, size=6),
    "Years_of_Experience": np.random.randint(1, 31, size=6)
}
```

```python
df = pd.DataFrame(data)
```

```python
print("=" * 50)
print("Employee Dataset")
print("=" * 50)
print(df)
```

```python
print("\n==============================")
print("First 3 Rows")
print("==============================")
print(df.head(3))
```

```python
print("\n==============================")
print("Total Number of Employees")
print("==============================")
print(len(df))
```

```python
print("\n==============================")
print("Dataset Information")
print("==============================")
df.info()
```

```python
print("\n==============================")
print("Shape of Dataset")
print("==============================")
print(df.shape)
```

```python
print("\n==============================")
print("Statistical Summary")
print("==============================")
print(df.describe())
```

```python
print("\n==============================")
print("Salary Statistics")
print("==============================")

print("Mean Salary:")
print(df["Salary"].mean())

print("\nMedian Salary:")
print(df["Salary"].median())

print("\nMode Salary:")
print(df["Salary"].mode())
```

```python
print("\n==============================")
print("Experience Statistics")
print("==============================")

print("Variance:")
print(df["Years_of_Experience"].var())

print("\nStandard Deviation:")
print(df["Years_of_Experience"].std())
```

```python
average_salary = df["Salary"].mean()
```

```python
print("\n==============================")
print("Employees Earning Above Average Salary")
print("==============================")

above_avg = df[df["Salary"] > average_salary]
print(above_avg)

print("\n==============================")
print("Employees with Experience >=5 and Salary <50000")
print("==============================")
```

```python
filtered = df[
    (df["Years_of_Experience"] >= 5) &
    (df["Salary"] < 50000)
]
```

```python
print(filtered)
```

```python
print("\n==============================")
print("Employees Sorted by Age (Descending)")
print("==============================")

sorted_df = df.sort_values(by="Age", ascending=False)

print(sorted_df)

```

```python
Q1 = df["Salary"].quantile(0.25)
Q2 = df["Salary"].quantile(0.50)
Q3 = df["Salary"].quantile(0.75)

IQR = Q3 - Q1
```

```python
print("\n==============================")
print("Quartiles")
print("==============================")

print("Q1 =", Q1)
print("Q2 =", Q2)
print("Q3 =", Q3)

print("\nIQR =", IQR)
```

```python
lower_limit = Q1 - (1.5 * IQR)
upper_limit = Q3 + (1.5 * IQR)

print("\nLower Limit =", lower_limit)
print("Upper Limit =", upper_limit)

outliers = df[
    (df["Salary"] < lower_limit) |
    (df["Salary"] > upper_limit)
]
```

```python
print("\n==============================")
print("Salary Outliers")
print("==============================")
print(outliers)
```

```python
print("\n==============================")
print("Correlation (Age & Salary)")
print("==============================")

correlation = df[["Age", "Salary"]].corr()

print(correlation)
```

```python
print("\n==============================")
print("Employee(s) with Highest Salary")
print("==============================")

highest_salary = df["Salary"].max()

highest_paid = df[df["Salary"] == highest_salary]

print(highest_paid)

print("\n==============================")
print("Youngest Employee with Highest Experience")
print("==============================")

```

```python
max_exp = df["Years_of_Experience"].max()

highest_exp = df[
    df["Years_of_Experience"] == max_exp
]

youngest_highest_exp = highest_exp.sort_values(
    by="Age",
    ascending=True
).head(1)

print(youngest_highest_exp)
```

```python
plt.figure(figsize=(6,4))

plt.hist(
    df["Salary"],
    bins=5,
    edgecolor="black"
)

plt.title("Salary Distribution")
plt.xlabel("Salary")
plt.ylabel("Frequency")

plt.show()
```

```python
plt.figure(figsize=(6,4))

sns.boxplot(
    y=df["Years_of_Experience"]
)

plt.title("Years of Experience")

plt.show()
```

```python
plt.figure(figsize=(6,4))

plt.scatter(
    df["Years_of_Experience"],
    df["Salary"]
)

plt.title("Salary vs Years of Experience")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")

plt.show()
```

```python
sns.pairplot(
    df[
        [
            "Age",
            "Salary",
            "Years_of_Experience"
        ]
    ]
)

plt.show()
```

```python
corr = df[
    [
        "Age",
        "Salary",
        "Years_of_Experience"
    ]
].corr()

plt.figure(figsize=(6,4))

sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Heatmap")

plt.show()

print("\nProgram Completed Successfully!")

```
